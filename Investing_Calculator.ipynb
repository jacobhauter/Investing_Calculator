```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def calculate_investment(principal, rate, years):
    return principal * ((1 + rate) ** years - 1) / rate

def currency_formatter(x, pos):
    return "${:,.0f}".format(x)

def main():
    print("Retirement Forecasting Calculator\n")

#     current_age = int(input("Enter your current age: "))
    current_age = 30
#     retirement_age = int(input("Enter your planned retirement age: "))
    retirement_age = 55
    
    # Calculate the number of years until retirement
    years_until_retirement = retirement_age - current_age

    # Input current savings for IRA, 401k, and HSA, and their respective annual return rates
#     current_ira_savings = float(input("Enter your current IRA savings: $"))
    current_ira_savings = 40000
#     current_ira_savings_return_rate = float(input("Enter the annual return rate for your current IRA savings (as a decimal): "))
    current_ira_savings_return_rate = .07
    
#     current_401k_savings = float(input("Enter your current 401k savings: $"))
    current_401k_savings = 45000
#     current_401k_savings_return_rate = float(input("Enter the annual return rate for your current 401k savings (as a decimal): "))
    current_401k_savings_return_rate = .07
    
#     current_hsa_savings = float(input("Enter your current HSA savings: $"))
    current_hsa_savings = 0
#     current_hsa_savings_return_rate = float(input("Enter the annual return rate for your current HSA savings (as a decimal): "))
    current_hsa_savings_return_rate = .07
    
#     annual_investment_ira = float(input("Enter annual investment amount for IRA: $"))
    annual_investment_ira = 6500
#     annual_investment_401k = float(input("Enter annual investment amount for 401k: $"))
    annual_investment_401k = 15000
#     annual_investment_hsa = float(input("Enter annual investment amount for HSA: $"))
    annual_investment_hsa = 3500
    
#     annual_return_ira = float(input("Enter annual return rate for IRA (as a decimal): "))
    annual_return_ira = .07
#     annual_return_401k = float(input("Enter annual return rate for 401k (as a decimal): "))
    annual_return_401k = .07
#     annual_return_hsa = float(input("Enter annual return rate for HSA (as a decimal): "))
    annual_return_hsa = .07

    combined_values = []
    ages = []  # To store the corresponding ages
    for year in range(years_until_retirement + 1):
        ira_value = calculate_investment(annual_investment_ira, annual_return_ira, year)
        _401k_value = calculate_investment(annual_investment_401k, annual_return_401k, year)
        hsa_value = calculate_investment(annual_investment_hsa, annual_return_hsa, year)
        
        # Calculate the annual return on the current savings for IRA, 401k, and HSA
        current_ira_savings_return = current_ira_savings * (1 + current_ira_savings_return_rate)
        current_401k_savings_return = current_401k_savings * (1 + current_401k_savings_return_rate)
        current_hsa_savings_return = current_hsa_savings * (1 + current_hsa_savings_return_rate)
        
        combined_value = (
            ira_value + _401k_value + hsa_value +
            current_ira_savings_return + current_401k_savings_return + current_hsa_savings_return
        )
        combined_values.append(combined_value)
        ages.append(current_age + year)  # Calculate the age for each year

    # Plot the combined investment values over time with age on the x-axis
    plt.figure(figsize=(10, 6))
    plt.plot(ages, combined_values, label='Combined Accounts', marker='o', color='blue')
    plt.title("Financial Investment Forecast")
    plt.xlabel("Age")
    plt.ylabel("Investment Value")
    plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(currency_formatter))
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    main()
```
